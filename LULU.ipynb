```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lulu = pd.read_csv('./LULU.csv')
lulu
```

```python
# check the dtype
lulu.info()
```

```python
# draw lineplot using seaborn
plt.figure(figsize=(20,10))
plt.title("Lululemon Historical Data", size=18)
plt.xlabel('Months', size=15)
plt.ylabel('Stock Price', size=15)
plt.ylim(250, 500)
plt.xticks(rotation=30)
sns.set_theme(style='whitegrid')
sns.lineplot(data=lulu, x='Date', y='Close')
```

```python
# draw barplot using seaborn
plt.figure(figsize=(20,10))
plt.title("Lululemon Historical Data", size=18)
plt.xlabel('Months', size=15)
plt.ylabel('Stock Price', size=15)
plt.ylim(250, 500)
plt.xticks(rotation=35)
ax = sns.barplot(data=lulu, x='Date', y='Close')
ax.set(ylabel='Stock Price', xlabel='Months')
```
